```python
import numpy as np
import pandas as pd
```

```python
filepath = '/content/drive/MyDrive/WA_Fn-UseC_-Telco-Customer-Churn.csv'
df = pd.read_csv(filepath)
```

```python
df.head()
```

```python
df.columns
```

```python
df.dtypes
```

```python
df.drop(df.columns[[0]], axis=1, inplace = True)
```

```python
df.drop(['customerID'], axis=1, inplace = True)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df['TotalCharges'][488]
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
# Replace 'Yes' with 1 and 'No' with 0 in the 'ColumnName' column
df['Churn'] = df['Churn'].replace({'Yes': 1, 'No': 0})
```

```python
df.head()
```

```python
dataframe = pd.get_dummies(df)
```

```python
dataframe.columns.values
```

```python
print(len(dataframe),len(dataframe.columns))
```

```python
dataframe.head()
```

```python
Y = dataframe['Churn'].values
X = dataframe.drop(['Churn'], axis=1)
```

```python
X
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler(feature_range = (0,1))
model=scaler.fit(X)
X = model.transform(X)
```

```python
X
```

```python
# Create Train & Test Data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

```python
len(Y_test)
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel='linear')
```

```python
model.fit(X_train,Y_train)
```

```python
predictions = model.predict(X_test)
```

```python
# Print the prediction accuracy
accuracy = metrics.accuracy_score(Y_test, predictions)
print("Accuracy:", accuracy)
print(" ")
print("Confusion_Matrix")
print(confusion_matrix(Y_test,predictions))
print(" ")
# Generate a classification report
class_report = classification_report(Y_test, predictions)
print(" ")
print("Classification Report")
print(class_report)
```
